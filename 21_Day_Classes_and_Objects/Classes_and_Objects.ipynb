##### 1. Python has the module called statistics and we can use this module to do all the statistical calculations. However, to learn how to make function and reuse function let us try to develop a program, which calculates the measure of central tendency of a sample (mean, median, mode) and measure of variability (range, variance, standard deviation). In addition to those measures, find the min, max, count, percentile, and frequency distribution of the sample. You can create a class called Statistics and create all the functions that do statistical calculations as methods for the Statistics class. Check the output below.

```python
import statistics
import numpy as np

class Statistics:
    def __init__(self, input):
        self.input = input

    def find_mean(self):
        return statistics.mean(self.input)

    def find_median(self):
        return statistics.median(self.input)

    def find_mode(self):
        try:
            return statistics.mode(self.input)
        except statistics.StatisticsError:
            return "No mode found."

    def find_range(self):
        return max(self.input) - min(self.input)

    def find_variance(self):
        return statistics.variance(self.input)

    def find_standard_deviation(self):
        return statistics.stdev(self.input)

    def find_min(self):
        return min(self.input)

    def find_max(self):
        return max(self.input)

    def find_count(self):
        return len(self.input)

    def find_percentile(self, percentile):
        return np.percentile(self.input, percentile)

    def find_frequency_distribution(self):
        frequency_dict = {}
        for value in self.input:
            if value in frequency_dict:
                frequency_dict[value] += 1
            else:
                frequency_dict[value] = 1
        return frequency_dict

#given data
ages = [31, 26, 34, 37, 27, 26, 32, 32, 26, 27, 27, 24, 32, 33, 27, 25, 26, 38, 37, 31, 34, 24, 33, 29, 26]

# Creating an instance for the class Statistics
stats_cal = Statistics(ages)

# Calculations
print("Mean:", stats_cal.find_mean())
print("Median:", stats_cal.find_median())
print("Mode:", stats_cal.find_mode())
print("Range:", stats_cal.find_range())
print("Variance:", stats_cal.find_variance())
print("Standard Deviation:", stats_cal.find_standard_deviation())
print("Min:", stats_cal.find_min())
print("Max:", stats_cal.find_max())
print("Count:", stats_cal.find_count())
print("25th Percentile:", stats_cal.find_percentile(25))
print("75th Percentile:", stats_cal.find_percentile(75))
print("Frequency Distribution:", stats_cal.find_frequency_distribution())
```

#### 2. Create a class called PersonAccount. It has firstname, lastname, incomes, expenses properties and it has total_income, total_expense, account_info, add_income, add_expense and account_balance methods. Incomes is a set of incomes and its description. The same goes for expenses.


```python
class PersonAccount:
    def __init__(self, firstname, lastname):
        self.firstname = firstname
        self.lastname = lastname
        self.incomes = {}
        self.expenses = {}

    def add_income(self, amount, description):
        if description in self.incomes:
            self.incomes[description] += amount
        else:
            self.incomes[description] = amount

    def add_expense(self, amount, description):
        if description in self.expenses:
            self.expenses[description] += amount
        else:
            self.expenses[description] = amount

    def total_income(self):
        return sum(self.incomes.values())

    def total_expense(self):
        return sum(self.expenses.values())

    def account_balance(self):
        return self.total_income() - self.total_expense()

    def account_info(self):
        info = f"Account Information for {self.firstname} {self.lastname}:\n"
        info += f"Total Income: {self.total_income()}\n"
        info += f"Total Expense: {self.total_expense()}\n"
        info += f"Account Balance: {self.account_balance()}\n"
        info += "Incomes:\n"
        for description, amount in self.incomes.items():
            info += f"  {description}: {amount}\n"
        info += "Expenses:\n"
        for description, amount in self.expenses.items():
            info += f"  {description}: {amount}\n"
        return info

#Checking
person = PersonAccount("Hashim", "Adamu")
person.add_income(500, "Salary")
person.add_income(200, "Freelance Work")
person.add_expense(200, "Rent")
person.add_expense(100, "Utilities")


print(person.account_info())
```

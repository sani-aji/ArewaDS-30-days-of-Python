```python
#7 Quadratic equation is calculated as follows: ax² + bx + c = 0. Write a function which calculates solution set of a quadratic equation, solve_quadratic_eqn.
def quadratic_solver(a,b,c):
    descriminant = (((b**2 - 4*a*c)**0.5))
    x1 =(-b + descriminant)/(2*a)
    x2 = (-b - descriminant)/(2*a)
    return x1,x2
```

```python
#9 Declare a function named reverse_list. It takes an array as a parameter and it returns the reverse of the array (use loops)
def reverse_list(my_list):
    reversed_list = []
    for i in range(len(my_list) - 1, -1, -1):
        reversed_list.append(my_list[i])
    return reversed_list

# Trying
original_list = [1, 2, 3, 4, 5]
result = reverse_list(original_list)
print(f"Original List: {original_list}")
print(f"Reversed List: {result}")






```

```python
#Level3
#1 Write a function called is_prime, which checks if a number is prime.
def is_prime(x):
    if x< 2:
        return False
    for i in range(2, int(x**0.5) + 1):
        if x % i == 0:
            return False
    return True
```

```python
#3 Write a function which checks if all the items of the list are of the same data type.
def same_type(my_list):
    if not my_list:  #  If the list is empty, it returns True because an empty list is considered to have the same type here.
        return True 
    type1 = type(my_list[0])
    return all(type(item) == type1 for item in my_list)
list1 = [1, 2, 3, 4, 5]
list2 = [1, 2, 'three', 4, 5]
list3 = []

print(same_type(list1))  
print(same_type(list2))  
print(same_type(list3))  
```

```python
#4 Write a function which check if provided variable is a valid python variable
def is_valid_variable(var_name):
    # Check if the variable name is a valid identifier
    if not var_name.isidentifier():
        return False

    # Check if the variable name is not a Python keyword
    import keyword
    if keyword.iskeyword(var_name):
        return False

    # Check if the variable name does not start with a digit
    if var_name[0].isdigit():
        return False

    return True
```

```python
print(is_valid_variable("valid_variable")) 
print(is_valid_variable("1_invalid_variable"))  
print(is_valid_variable("class"))  
```
